```python
DNA_to_AA = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M','ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K','AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L','CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q','CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V','GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E','GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S','TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_','TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W',
    }
# This defines the dictionary with Amino acids as values and DNA codons as keys, underscores represent stop codons
```

```python
file = open("DNA_sequence.txt")
DNA_sequence = file.read()
DNA_sequence = DNA_sequence.upper() # converts the characters to upper case as this is the case of the dictionary
# This creates the object to be tranlated
```

```python
t = str.maketrans({"A": "T", "G": "C" , "T":"A", "C":"G"}) #creates a translation table for the completmentary/reverse strand
DNA_sequence_t = DNA_sequence.translate(t) #translates the sequence to make the completementary DNA strand
```

```python
DNA_sequence_RF0 = DNA_sequence[0:]
DNA_sequence_RF1 = DNA_sequence[1:]
DNA_sequence_RF2 = DNA_sequence[2:] # creates new strings starting from first, second or third character
DNA_sequence_RF0_rev = DNA_sequence_t[::-1] 
DNA_sequence_RF1_rev = DNA_sequence_t[-2::-1]
DNA_sequence_RF2_rev = DNA_sequence_t[-3::-1] # creates new str starting from first, second or third character of comp strand

reading_frames = { #creates a dictionary of labels for the objects in my reading_frames, RF= reading frame, rev= reverse
    "DNA_sequence_RF0": DNA_sequence_RF0,
    "DNA_sequence_RF1": DNA_sequence_RF1,
    "DNA_sequence_RF2": DNA_sequence_RF2,
    "DNA_sequence_RF0_rev": DNA_sequence_RF0_rev,
    "DNA_sequence_RF1_rev": DNA_sequence_RF1_rev,
    "DNA_sequence_RF2_rev": DNA_sequence_RF2_rev,
}
```

```python
for label, obj in reading_frames.items(): # this loops through every item in the dictionary
    if len(obj)%3 == 0: # checks sequence in divisible by 3 to prevent a key error
        obj = obj # this is just a filler to let the loop work
    else:
        obj = obj[:-(len(obj)%3)] ## if sequence length not divisible by 3 removes the remainder
    
    AA_sequence = "" # This creates the empty string which will get filled up with translated amino acids
    for base in range(0, len(obj), 3): # This selects base as every 3 characters in the DNA sequence
        codon = obj[base:base + 3]  # Defines a codon as the characters between the steps of 3 and then moves to the next 3
        AA = DNA_to_AA[codon] # Reads the codon in the dictionary and translates to an amino acid
        AA_sequence+=(AA) # adds the amino acid to the amino acid sequence
    print(f"{label} translation: {AA_sequence}") ## prints the translation
```

```python
class color: ## colour codes to highlight restriction sites
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
restriction_sequence = "AATATT" # The recognition sequence for any restriciton enzyme can be entered here.
cut_sequence = f"{restriction_sequence[0:3]} //////// {restriction_sequence[3:6]}" # splits the restriction sequence at cut point

DNA_sequence_RF0_restriction = DNA_sequence_RF0.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF0 restriction sites" + color.END,  DNA_sequence_RF0_restriction)
DNA_sequence_RF1_restriction = DNA_sequence_RF1.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF1 restriction sites" + color.END,  DNA_sequence_RF1_restriction)
DNA_sequence_RF2_restriction = DNA_sequence_RF2.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF2 restriction sites" + color.END,  DNA_sequence_RF2_restriction)
DNA_sequence_RF0_rev_restriction = DNA_sequence_RF0_rev.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF0 reverse restriction sites" + color.END,  DNA_sequence_RF0_rev_restriction)
DNA_sequence_RF1_rev_restriction = DNA_sequence_RF1_rev.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF1 reverse restriction sites" + color.END,  DNA_sequence_RF1_rev_restriction)
DNA_sequence_RF2_rev_restriction = DNA_sequence_RF2_rev.replace(restriction_sequence, color.BOLD + color.RED + cut_sequence + color.END)
print(color.BOLD + "DNA sequence RF2 reverse restriction sites" + color.END,  DNA_sequence_RF2_rev_restriction)

#### this section looks for restriction sites within each of reading frames of the DNA sequence
#### The code identifies these sites with slashes and red text
```
